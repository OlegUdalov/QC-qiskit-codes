# In this Notebook I just play with statevectors


```python
import qiskit
from qiskit import QuantumRegister as Q_R
from qiskit import ClassicalRegister as C_R
from qiskit_aer import Aer
from qiskit_aer import AerSimulator
from qiskit import QuantumCircuit, transpile
```

```python
qr = Q_R(4)
cr = C_R(1)

circ = qiskit.QuantumCircuit(qr, cr)

circ.h(0)
#circ.x(1)

SimulatorAer = AerSimulator()

circ.save_statevector()

circ.draw('mpl')
```

```python
from qiskit.quantum_info import partial_trace, DensityMatrix
from qiskit.visualization import plot_state_city

circ = transpile(circ, backend = SimulatorAer)
result = SimulatorAer.run(circ,shots = 1).result()
psi = result.get_statevector(circ)
print(psi)

dm = DensityMatrix(psi)

q1 = partial_trace(dm, [1,2,3])

psi.draw(output = 'latex')

plot_state_city(q1)
```
